```python
from bs4 import BeautifulSoup
import requests
```

```python
page_link = 'https://loco2.com/en-us/journey/paris-chamonix-1iqtqs1?currency=USD&outbound=undefined&leg=outbound'
```

```python
# Fetch content from URL

page_response = requests.get(page_link, timeout=5)
```

```python
page_response
```

```python
# Parse HTML

page_content = BeautifulSoup(page_response.content, "html.parser")
```

```python
prices = page_content.find_all(class_='price')
```

```python
prices
```

```python
prices = page_content.find_all('div', attrs={'class':'price'})
```
